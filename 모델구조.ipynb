
## * 모델의 layer들이 생성되는 원리에 대해 정리


```
from keras import models
from keras import layers
import numpy as np

x_train=np.array([1,2,3,4,5,])
y_train=np.array([1,2,3,4,5,])
x_test=np.array([11,12,13,14,15])
```

```
model=models.Sequential()
model.add(layers.Dense(5, activation='relu', input_shape=(1,)))
model.add(layers.Dense(3))
model.add(layers.Dense(1))

model.summary()
```

image.png



*   model=models.Sequential()

    : **레이어들을 선형으로 연결하여 구성한다.**

 

*   model.add(layers.Dense(5, activation='relu', input_shape=(1,)))

    : **첫번째 레이어는 입력 데이터 형태에 대한 정보(input_shape)을 받아야 한다.**

    : 5개의 노드를 가진 Dense 레이어를 통해 연결한다.


*   model.add(layers.Dense(3)) : 3개의 노드를 가진 Dense 레이어를 통해 연결한다.


*   model.add(layers.Dense(1)) : 하나의 값을 출력하는 마지막 노드이다.

---




*   Output Shape (연결된 레이어의 노드 수와 형태를 알 수 있다.)



    (None,5)

    (None,3)

    (None,1)




*   Parameter (해당 가중치와 bias 개수의 총합을 의마한다.)


    10

    18  

     4

* Output Shape에서 행이 None으로 표시되는 이유 :

 → 데이터의 개수는 계속해서 추가될 수 있기 때문에 딥러닝 모델에서는 행을 무시하고 열의 shape을 맞춰주는 작업을 많이 수행한다.

```
model.compile(loss='mse',                           #모델을 학습하기 전에, compile을 통해서 학습 방식에 대한 환경설정을 한다.
              optimizer='adam')

model.fit(x_train,y_train,epochs=10,batch_size=2)   #fit을 통해서 입력데이터와 라벨을 가지고 학습을 진행한다.

mse=model.evaluate(x_train, y_train, batch_size=2)
print('mse: ', mse)
```
