
4. **(10pts)** Given any random matrices, compute the element-wise multiplication using a naive Python implementation and Numpy built-in function respectively. Compare the wall-clock times of these implementations as the size of matrices increases.

````
#multiplication using naive Python implementation

import numpy as np
import time

start = time.time()

def naive_vector_dot(x,y):
  assert len(x.shape)==1
  assert len(y.shape)==1
  assert x.shape[0]==y.shape[0]

  z=0
  for i in range(x.shape[0]):
    z += x[i] * y[i]
  return z

def naive_matrix_vector_dot(m,n,r):
  x = np.random.randint(10,size=(m,n))
  y = np.random.randint(10,size=(n,r))
  assert len(x.shape) == 2
  assert len(y.shape) == 2
  assert x.shape[1] == y.shape[0]

  z=np.zeros((x.shape[0], y.shape[1]))
  for i in range(x.shape[0]):
    for j in range(y.shape[1]):
      row_x=x[i, :]
      column_y=y[:,j]
      z[i,j]=naive_vector_dot(row_x, column_y)
  return z


A=naive_matrix_vector_dot(10,50,15)
print(A)
print('the time of multiply (10 by 50) and (50 by 15) matrices : %.4f' % (time.time()-start))

B=naive_matrix_vector_dot(100,500,150)
print(B)
print('the time of multiply (100 by 500) and (500 by 150) matrices : %.4f' % (time.time()-start))
````

naive Python 으로 했을 때

(10 by 50), (50 by 15) 매트릭스를 연산하면 0.0058초가 걸리고,

(100 by 500), (500 by 150) 매트릭스를 연산하면 3.5267초가 걸린다는 것을 알 수 있다.

즉, 행렬의 크기가 커질수록 결과를 내는 데에 시간이 더 많이 소요된다.




````
#multiplication using Numpy built-in function

import numpy as np
import time

start=time.time()

def naive_matrix_vector_dot(m,n,r):
  x = np.random.randint(10,size=(m,n))
  y = np.random.randint(10,size=(n,r))
  z=np.dot(x,y) # 행렬의 곱셈 연산을 하려면 x.shape[1] == y.shape[0] 이어야 한다.
  return z

A=naive_matrix_vector_dot(10,50,15)
print(A)
print('the time of multiply (10 by 50) and (50 by 15) matrices : %.4f' % (time.time()-start))

B=naive_matrix_vector_dot(100,500,150)
print(B)
print('the time of multiply (100 by 500) and (500 by 150) matrices : %.4f' % (time.time()-start))
````

Numpy built-in function 으로 했을 때

(10 by 50), (50 by 15) 매트릭스를 연산하면 0.0023초가 걸리고,

(100 by 500), (500 by 150) 매트릭스를 연산하면 0.0187초가 걸린다는 것을 알 수 있다.

결국 크기가 큰 매트릭스의 연산에 소요되는 시간을 비교 해보면, Numpy built-in function이 naive Python에 비해 훨씬 더 빠르게 처리된다는 것을 알 수 있다.



5. **(10pts)** Consider MNIST classification problem covered during the class. For the details, please refer to [this material](https://drive.google.com/a/ds.seoultech.ac.kr/file/d/1lWLwZQGXgZJq4PFHvgra6D6U420ra-O3/view?usp=sharing). In this example, we used the multilayer perceptron composed of an input layer with 512 hidden nodes and an output layer that produces predicted probabilities over 10 classes. In the class, we used GPU as a hardware accelerator to train our model.

  Here, let's verify the actual benefit of using GPU for training. For this, compare the wall-clock times in the case of 1) using CPU and 2) using GPU to train MNIST classifier.

````
import tensorflow as tf
tf.__version__

import time

from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

from tensorflow.keras import models
from tensorflow.keras import layers

network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
network.add(layers.Dense(10, activation='softmax'))

network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

train_images=train_images.reshape((60000, 28*28))
train_images=train_images.astype('float32') / 255

test_images=test_images.reshape((10000, 28*28))
test_images=test_images.astype('float32') / 255

from tensorflow.keras.utils import to_categorical

train_labels=to_categorical(train_labels)
test_labels=to_categorical(test_labels)

start=time.time()

with tf.device('/CPU:0'):
  nf=network.fit(train_images, train_labels, epochs=5, batch_size=128)

run_time_cpu = time.time()-start
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc: ', test_acc)
print('Elapsed time of using CPU: %.4f' % (run_time_cpu))

start1=time.time()

with tf.device('/GPU:0'):
  nf=network.fit(train_images, train_labels, epochs=5, batch_size=128)

run_time_gpu = time.time()-start1
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc: ', test_acc)
print('Elapsed time of using GPU: %.4f' % (run_time_gpu))

print('GPU speedup over CPU: %.4f'%(run_time_cpu/run_time_gpu))
````

with tf.device 로 연산을 수행할 장치(CPU/GPU)를 직접 선택한다.

'Elapsed time of using CPU: 17.4259초

'Elapsed time of using GPU: 8.8967초

위의 코드 실행결과로, GPU로 실행했을 때가 CPU로 실행했을 때보다 1.9587배 더 빠르게 처리된다는 것을 알 수 있다.
