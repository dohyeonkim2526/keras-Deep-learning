# 케라스를 사용한 단어 수준의 원-핫 인코딩



*   keras.prerocessing.text.Tokenizer(텍스트 토큰화 유틸리티 클래스)

  → 각 텍스트를 정수 시퀀스 또는 이진 벡터로 변환하여 말뭉치(corpus)를 벡터화할 수 있도록 해준다.

          keras.preprocessing.text.Tokenizer(num_words=None, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True, split=' ')

  num_words: 가장 빈도가 높은 (num_words)개의 단어만 선택한다.

  filters: ('문자는 제외)텍스트에서 걸러진 문자

  lower: 텍스트를 소문자로 변환할지에 대한 여부

  split: 문자열을 단어로 분해해주는 역할




## [ 빈도가 높은 10개의 단어를 가지고 **단어 인덱스 구축**하기 ]

  **from** keras.preprocessing.text **import** Tokenizer

  samples=['The cat sat on the mat.', 'The dog ate my homework.']

  tokenizer=Tokenizer(num_words=10)

  tokenizer.**fit_on_texts**(samples)

  word_index=tokenizer.**word_index**

  sequences=tokenizer.**texts_to_sequences**(samples)

  one_hot_results=tokenizer.**texts_to_matrix**(samples, mode='binary')


```
from keras.preprocessing.text import Tokenizer

samples=['The cat sat on the mat.', 'The dog ate my homework.']

tokenizer=Tokenizer(num_words=10)  # 가장 빈도가 높은 10개의 단어를 선택하여 Tokenizer 객체를 만든다.
```

```
tokenizer.fit_on_texts(samples)   # 단어 인덱스 구축
```

```
word_index=tokenizer.word_index
print(word_index)
```

## [ 문장을 단어 인덱스로 변환해서 **원-핫 인코딩** 하기 ]


    'The cat sat on the mat.' → The(1), cat(2), sat(3), on(4), the(1), mat(5)
    'The dog ate my homework.' → The(1), dog(6), ate(7), my(8), homework(9) 

```
sequences=tokenizer.texts_to_sequences(samples)
print(sequences)
```

```
one_hot_results=tokenizer.texts_to_matrix(samples, mode='binary')
print(one_hot_results)
```

```
print('%s개의 고유한 토큰을 찾았습니다.' % len(word_index))
```
