## º과소적합(Underfitting)
- test set의 성능이 향상될 여지가 아직 있을 때 발생한다.
- 즉, 네트워크가 train set에서 적절한 패털을 학습하지 못했음을 의미한다.
- 원인: 모델이 너무 단순하다 / 규제가 너무 많다 / 단순히 충분히 오래 훈련하지 않았다.


## º과대적합(Overfitting)
- train set에 특화된 패턴을 학습하기 시작했을 때 발생한다.
- 모델이 너무 오래 훈련되었을 때 발생하므로, (과소적합과 과대적합을 모두 해결할 수 있는)적절한 epoch동안 학습을 진행해야 한다.


## [과대적합을 해결할 수 있는 규제(Regularization)]
## 1.가중치 규제
- 간단한 모델이 복잡한 모델보다 과대적합이 될 가능성이 낮다.
- 그러므로 '가중치 규제'를 통해, 네트워크의 복잡도에 제한을 두어 가중치가 작은 값을 가지도록 만든다.
- 네트워크의 loss function에 큰 가중치에 연관된 비용(cost)을 추가한다.

      1) L1 regularization: 가중치의 절댓값에 비례하는 비용이 추가된다.
      2) L2 regularization: 가중치의 제곱에 비례하는 비용이 추가된다. 

      model.add(layers.Dense(16, kernel_regularizer=regularizers.l1(0.001), activation='relu', input_shape=(10000,))) ... L1규제
      model.add(layers.Dense(16, kernel_regularizer=regularizers.l2(0.001), activation='relu', input_shape=(10000,))) ... L2규제
      model.add(layers.Dense(16, kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.001), activation='relu', input_shape=(10000,))) ... L1과 L2규제 병행



## 2.드롭아웃(dropout)
- 훈련하는 동안 무작위로 층의 일부 출력 특성을 제외시킨다.
- 드롭아웃의 비율은 0이 될 특성의 비율을 가리킨다.
- 드롭아웃을 하면, 가중치들 사이에 동조화(co-adaptation: 비슷한 특징에 집중함) 되는 현상을 피할 수 있게 된다.

      model.add(layers.Dropout(0.5))
